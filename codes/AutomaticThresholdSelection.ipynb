```python
from plantcv import plantcv as pcv  
import cv2
import numpy as np 
%matplotlib inline
pcv.params.debug = "print"
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../../thumbnail_Exp11-1.jpg')
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
pcv.plot_image(img)
```

```python
threshold_otsu_light  = pcv.threshold.otsu(gray_img=gray_img, max_value=255, object_type="light")
threshold_gaussian = pcv.threshold.gaussian(gray_img=gray_img, max_value=255, object_type='dark')
threshold_binary_light = pcv.threshold.binary(gray_img=gray_img, threshold=36, max_value=255, object_type='light')
threshold_mean = pcv.threshold.mean(gray_img=gray_img, max_value=255, object_type='dark')
sat_thresh = pcv.threshold.saturation(rgb_img=image, threshold=250, channel="any")
threshold_triangle = pcv.threshold.triangle(gray_img=gray_img, max_value=255, object_type='light', xstep=10)
```

```python
(
np.sum(threshold_otsu_light), 
 np.sum(threshold_gaussian), 
 np.sum(threshold_binary_light),
np.sum(threshold_mean),
np.sum(sat_thresh),
    np.sum(threshold_triangle)
)
```
